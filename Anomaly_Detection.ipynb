```python
#Encoding character variables as ml learns in numeric terms
from sklearn.preprocessing import LabelEncoder
before=df[['protocol_type','service','flag','class']]
for col in df.columns:
   if df[col].dtype=="object":
    le_col=LabelEncoder()
    df[col]=le_col.fit_transform(df[col])
after=df[['protocol_type','service','flag','class']]
```

```python
#built a dictionary of labels before and after encoding so that it is easier to identify
p=before.protocol_type.unique()
p1=[ df.loc[df['protocol_type']==i].index[0] for i in df['protocol_type'].unique()]
d1={}
for i in range(len(p)):
    d1[p[i]]=after.loc[p1[i]]['protocol_type']
print(d1)

q=before.service.unique()
p2=[df.loc[df['service']==i].index[0] for i in df['service'].unique()]
d2={}
for i in range(len(q)):
    d2[q[i]]=after.loc[p2[i]]['service']
print(d2)


r=before.flag.unique()
p3=[df.loc[df['flag']==i].index[0] for i in df['flag'].unique()]
d3={}
for i in range(len(r)):
    d3[r[i]]=after.loc[p3[i]]['flag']
print(d3)


s=before['class'].unique()
p4=[df.loc[df['class']==i].index[0] for i in df['class'].unique()]
d4={}
for i in range(len(s)):
    d4[s[i]]=after.loc[p4[i]]['class']
print(d4)
```

```python
#feature selection using mutual info(how much info can be extracted from one r.v given another)
from sklearn.feature_selection import mutual_info_classif
X = df.drop(['class'],axis=1)
y=df['class']
mutual_info=mutual_info_classif(X,y)
mutual_info
```

```python
#we analyse the mutual info recieved and plot a graph in descending order for noise reduction
mutual_info = pd.Series(mutual_info)
mutual_info.index=X.columns
mutual_info=mutual_info.sort_values(ascending=False)
mutual_info.plot.bar(figsize=(10,5))
```

```python
#picked best 26 features and reduced data
from sklearn.feature_selection import SelectKBest
best_cols=SelectKBest(mutual_info_classif,k=26)
print(best_cols)
```

```python
#print features
best_cols.fit(X,y)
selected_features = X.columns[best_cols.get_support()]
print(selected_features)
```

```python
#modify the dataset according to selected features
X=X[selected_features]
```

```python
#splitting of dataset for preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
#scale the data for better comparison of features and improved performance
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
#print
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
#using random forest classifier , we calculated the training time for the model
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

import time

clfr = RandomForestClassifier(n_estimators = 30)
start = time.time()

clfr.fit(X_train, y_train.values.ravel())

end = time.time()
     
print("Training time of RandomForestClassifier: ", end-start)
```

```python
#evaluating model performance
print("Train score is:", clfr.score(X_train, y_train))
print("Test score is:", clfr.score(X_test, y_test))
```

```python
#using naive bayes classfier , we calculated the training time for the model
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

clfg = GaussianNB()
start = time.time()
clfg.fit(X_train, y_train)
end = time.time()
print("Training time of Navie Bayes: ", end-start)
```

```python
#evaluating model performance
print("Train score is:", clfg.score(X_train, y_train))
print("Test score is:", clfg.score(X_test, y_test))
```

```python
#using decision tree , we calculated the training time for the model
from sklearn.tree import DecisionTreeClassifier

clfd = DecisionTreeClassifier(criterion ="entropy", max_depth = 4)
start = time.time()

clfd.fit(X_train, y_train)
end = time.time()
print("Training time of Decision Tree: ", end-start)
```

```python
#evaluating model performance
print("Train score is:", clfd.score(X_train, y_train))
print("Test score is:", clfd.score(X_test, y_test))
```

```python
#using decision tree , we calculated the training time for the model
from sklearn.linear_model import LogisticRegression

clfl = LogisticRegression(max_iter = 1200000)
start = time.time()

clfl.fit(X_train, y_train)
end = time.time()
print("Training time of Logistic Regression: ", end-start)
```

```python
#evaluating model performance
print("Train score is:", clfl.score(X_train, y_train))
print("Test score is:", clfl.score(X_test, y_test))
```
